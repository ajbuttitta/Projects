```python
df4 = df3.drop(columns=['CPI'])
df5 = df3["CPI"]


from linearmodels import PanelOLS
from linearmodels import RandomEffects
import statsmodels.api as sm

exog = sm.tools.tools.add_constant(df4)

# random effects model
model_re = RandomEffects(df4, df5) 
re_res = model_re.fit() 

# fixed effects model
model_fe = PanelOLS(df4, df5, entity_effects = True) 
fe_res = model_fe.fit() 

#print results
print(re_res)
print(fe_res)
```
