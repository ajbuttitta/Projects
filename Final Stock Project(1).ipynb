## Final Project- Stock Data compared to Google Trends

```python
import pytrends
import yfinance as yf

import requests
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import tkinter as tk
import tkinter.messagebox as mb
from pytrends.request import TrendReq
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import pandas_datareader as pdr
from matplotlib.figure import Figure
from tkinter import ttk

%matplotlib inline
```

```python
#Setup pytend
pytrend = TrendReq()

#Display functions
def showinfo(i):
    #recommendation stock data
    a= rec_data(i)
    a = a.tail(5)
    label4.config(text=a)
    #stock earnings data
    b = earnings_data(i)
    b = b.head(5)
    label5.config(text=b)
    #quarterly financials data
    c = qf_data(i)
    c = c.head(5)
    label6.config(text=c)
    #google data
    d = googledata(i)
    d = d.head(10)
    label8.config(text=d)
    
    #Setup Google Trends
    
    pytrend = TrendReq(hl='en-US', tz=360)
    keywords = [i]
    
    #Formatting the graphs
    pytrend.build_payload(kw_list=keywords,cat=0, timeframe='today 5-y', geo='US', gprop='')
    figure2 = plt.Figure(figsize=(4.6,3), dpi=100)
    ax2 = figure2.add_subplot(111)
    data2 = pytrend.interest_over_time()
    data2 = data2.drop(labels=['isPartial'],axis='columns')
    ax2.plot(data2[i], color = 'red')
    ax2.grid(True)
    ax2.set_title(i+ " over the past 5 years on Google Trends")    
    
    # Request data via Yahoo public API
    data = pdr.get_data_yahoo(i)
    figure = plt.Figure(figsize=(4.6,3), dpi=100)
    ax = figure.add_subplot(111)
    ax.plot(data['Adj Close'])
    ax.grid(True)
    ax.set_title(i+ " Stock Price over the past 5 years")
    
    #Formatting the graphs
    canvas = FigureCanvasTkAgg(figure, master=root)
    canvas.draw()
    canvas.get_tk_widget().pack(side=tk.LEFT)
    canvas2 = FigureCanvasTkAgg(figure2, master=root)
    canvas2.draw()
    canvas2.get_tk_widget().pack(side=tk.RIGHT)
    
#Get stock recommendations
def rec_data(i):
    stock = yf.Ticker(i)
    stock.history(period="5y")
    a=stock.recommendations
    return a

#Get stock earnings
def earnings_data(i):
    stock = yf.Ticker(i)
    stock.history(period="5y")
    b=stock.earnings
    return b

#Get quarterly financials
def qf_data(i):
    stock = yf.Ticker(i)
    stock.history(period="5y")
    c=stock.quarterly_financials
    return c

#Get GoogleTrend data and sort it so the top values are displayed
def googledata(i):
    kw_list = [i]
    pytrend.build_payload(kw_list, geo="US", timeframe='today 5-y')
    q = pytrend.interest_by_region(resolution='COUNTRY', inc_low_vol=True, inc_geo_code=False)
    q = q.sort_values(by=i, ascending=False)
    return q

    
#create a GUI window
root=tk.Tk()
root.geometry("1400x800")
root.title("Stock Analysis")

#Create labels
label1=tk.Label(root,text="Stock Analysis")
label1.config(font=('helvetica', 18))
label1.pack()
label=tk.Label(root,text="Enter stock:")
label.config(font=('helvetica', 10))
label.pack()

#take in user input
userinput=tk.StringVar(None)
entry=tk.Entry(root,textvariable=userinput)
entry.pack()

#Create a functional button
button=tk.Button(root,text="Click here to analyze",command=lambda: showinfo(userinput.get()), bg='blue', fg='white', font=('helvetica', 9, 'bold'))
button.pack()

#this part focuses on stock analysis
canvas1 = tk.Canvas(root, width = 600, height = 300,relief = 'raised')
canvas1.pack()

#label recommendations
label1 = tk.Label(root, text='Recommendations')
label1.config(font=('helvetica', 16))
canvas1.create_window(10, 60, window=label1)
#label4 = rec_data(i)
label4=tk.Label(root,text="")
label4.config(font=('helvetica',9))
canvas1.create_window(0,145, window=label4)

#label earnings
label2 = tk.Label(root, text='Earnings')
label2.config(font=('helvetica', 16))
canvas1.create_window(350, 60, window=label2)
#label5=earnings_data(i)
label5=tk.Label(root,text="")
label5.config(font=('helvetica',9))
canvas1.create_window(340,135, window=label5)

#label quarterly financials
label3 = tk.Label(root, text='Quarterly Financials')
label3.config(font=('helvetica', 16))
canvas1.create_window(300, 275, window=label3)
#label6 = qf_data(i)
label6=tk.Label(root,text="")
label6.config(font=('helvetica',9))
canvas1.create_window(305,400, window=label6)

#label interest by state
label7 = tk.Label(root, text='Interest by State')
label7.config(font=('helvetica', 16))
canvas1.create_window(615, 60, window=label7)
#label8=googledata(i)
label8=tk.Label(root,text="")
label8.config(font=('helvetica',9))
canvas1.create_window(615,175, window=label8)
```
